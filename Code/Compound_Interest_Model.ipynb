```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import pprint
```

```python
compound_interest_dict = []
mod5_dict = []

periods_list = []
beginning_vals = []
compound_interest_list = []

mod_5_beginning_vals = []
mod5_dict = []
mod5_periods_list = []
mod5_compound_interest_list = []

print("-- Time Value of Money --")
first_beginning_val = int(input("How much money (in U.S. dollars ($)) do you want to invest? "))
beginning_vals.append(int(first_beginning_val))

num_periods = int(input("How many years ? "))

interest_rate = int(input("What percentage is your interest rate? "))

simple_interest = round(interest_rate / 100 * first_beginning_val, 2)
ioi = round(0, 2)
compound_interest = simple_interest + ioi
end_value = first_beginning_val + compound_interest

for i in range(1, (int(num_periods) + 1)):
    if i == 1:
        first_period = {'Period' : i, 
                        'Beginning Value' : first_beginning_val,
                        'Simple Interest' : simple_interest,
                        'Interest on Interest' : ioi,
                        'Total (Compound Interest)' : compound_interest,
                        'End Value' : end_value
                       }
        compound_interest_dict.append(first_period)
        
        period = first_period.get('Period')
        periods_list.append(int(period))
        
        compound_interest = first_period.get('Total (Compound Interest)')
        compound_interest_list.append(int(compound_interest))
    else:
        beginning_val = compound_interest_dict[i - 2].get('End Value')
        beginning_vals.append(beginning_val)
        
        ioi = round(beginning_val * (interest_rate/100) - simple_interest, 2)
        compound_interest = round(simple_interest + ioi, 2)
        end_value = round(beginning_val + compound_interest, 2)
        
        periods = {
            'Period' : i,
            'Beginning Value' : beginning_val,
            'Simple Interest' : simple_interest,
            'Interest on Interest' : ioi,
            'Total (Compound Interest)' : compound_interest,
            'End Value' : end_value
        }
        
        compound_interest_dict.append(periods)
        
        period = periods.get('Period')
        periods_list.append(int(period))
        
        compound_interest = periods.get('Total (Compound Interest)')
        compound_interest_list.append(int(compound_interest))
        
        if i % 5 == 0:
            
            mod_5_beginning_val = compound_interest_dict[i - 2].get('End Value')
            mod_5_beginning_vals.append(int(mod_5_beginning_val))
            
            mod5_dict.append(periods)
            
            mod5_period = periods.get('Period')
            mod5_periods_list.append(int(mod5_period))
            
            mod5_compound_interest = periods.get('Total (Compound Interest)')
            mod5_compound_interest_list.append(int(mod5_compound_interest))


pprint.pprint(compound_interest_dict)
```

```python
c_df = pd.DataFrame(compound_interest_dict, columns = ['Period', 
                                                       'Beginning Value', 
                                                       'Simple Interest', 
                                                       'Interest on Interest', 
                                                       'Total (Compound Interest)',
                                                       'End Value']).set_index("Period")
c_df.tail()
```

```python
compound_df = pd.DataFrame(c_df, columns = ["Total (Compound Interest)"])
compound_df.tail()
```

```python
Total = compound_df['Total (Compound Interest)'].sum()
print(round(Total, 2))
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(16, 8))

ax.bar(periods_list[:num_periods], 
       beginning_vals[:num_periods],
       label = 'Beginning Value')
ax.bar(periods_list[:num_periods], 
       compound_interest_list[:num_periods],
       bottom = beginning_vals[:num_periods],
       label = 'Total (Compound Interest)')

ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_ylabel('End Values', fontsize=18)
ax.set_xlabel('Period', fontsize=18)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)  

fig.tight_layout()
fig.subplots_adjust(top=0.90)
fig.suptitle('Time Value of Money: ' + str(int(num_periods)) + ' years', fontsize="x-large", fontweight = "bold")
plt.show()
```

```python
plt.style.use('ggplot')

ax = rev_mod5_df.plot(stacked=True, kind='bar', figsize=(16, 8), rot='horizontal')

# .patches is everything inside of the chart
for rect in ax.patches:
    # Find where everything is located
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()
    
    # The height of the bar is the data value and can be used as the label
    label_text = f'${height:,.2f}'  # f'{height:.2f}' to format decimal values
    
    # ax.text(x, y, text)
    label_x = x + width / 2
    label_y = y + height / 2

    # plot only when height is greater than specified value
    if height > 0:
        ax.text(label_x, label_y, label_text, ha='center', va='center', fontsize=10)
    
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)    
ax.set_ylabel("Total Value", fontsize=18)
ax.set_xlabel("Period", fontsize=18)
plt.title('Time Value of Money: ' + str(int(num_periods)) + ' years', fontsize="x-large", fontweight = "bold")
plt.show()
```
